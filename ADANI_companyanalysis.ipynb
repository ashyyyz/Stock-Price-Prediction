```python
dataset.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer()
dataset['Trades'] = imputer.fit_transform(dataset.Trades.values.reshape(-1,1))
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()
dataset['Date'] = label_encode.fit_transform(dataset['Date'])
dataset['Symbol'] = label_encode.fit_transform(dataset['Symbol'])
dataset['Series'] = label_encode.fit_transform(dataset['Series'])
dataset
```

```python
dataset.info()
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Date")
plt.ylabel("close price")
dataset.plot.line(y="Close", x = "Date")
```

```python
data = dataset[["Close"]]
data
```

```python
data = data.rename(columns = {'Close':'Actual_Close'})
data
```

```python
dataset["Target"] = data.rolling(2).apply(lambda x: x.iloc[1] > x.iloc[0])["Actual_Close"]
```

```python
dataset["Target"]
```

```python
adani_prev = dataset.copy()
adani_prev = adani_prev.shift(1)
dataset
```

```python
adani_prev
```

```python
adani_prev["Target"][1] = 0.0
adani_prev
```

```python
data = data.join(adani_prev).iloc[1:]
data
```

```python
y = data["Target"]
y
```

```python
x = data.drop("Target", axis = 1)
```

```python
x
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components = 1)
lda_x = lda.fit_transform(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(lda_x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn import svm
model = svm.SVC() #SVC = suppport vector classifier -> svm has another one also that is SVR -> support vector regression also
# train the model

model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print("Accuracy Score : \n", accuracy_score(y_test, y_pred))
print("Confusion matrix : \n", confusion_matrix(y_test, y_pred))
print("Classification report : \n", classification_report(y_test, y_pred))
```
